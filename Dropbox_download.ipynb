```python
path_list = []

result = dbx.files_list_folder(path, limit=500)
for i in result.entries:
    if 'PUM' in i.name and 'SS' not in i.name and 'Sample' not in i.name:
        path_list.append(i.name)

while result.has_more:
    result = dbx.files_list_folder_continue(result.cursor)
    for i in result.entries:
        if 'PUM' in i.name and 'SS' not in i.name and 'Sample' not in i.name:
            path_list.append(i.name)

print(path_list)
print(len(path_list))
```
